```python
import pandas as pd

# 1. Load the file you just sent her
print("Loading the ADMET file...")
df = pd.read_csv('Encoded_Master_ADMET_Results.csv')

# 2. Define the exact rank for the solubility classes
solubility_map = {
    'Insoluble': 0,
    'Poorly soluble': 1,
    'Moderately soluble': 2,
    'Soluble': 3,
    'Very soluble': 4
}

# 3. Apply the encoding to the final three text columns
columns_to_encode = ['ESOL Class', 'Ali Class', 'Silicos-IT class']

print("Applying ordinal encoding to solubility columns...")
for col in columns_to_encode:
    if col in df.columns:
        # Map the text to the corresponding number
        df[col] = df[col].map(solubility_map).fillna(df[col])

# 4. Save the ultimate, 100% machine-learning-ready file
output_filename = 'Final_ML_Ready_ADMET.csv'
df.to_csv(output_filename, index=False)

print(f"\n✅ Success! Every single variable has now been encoded.")
print(f"File saved as: {output_filename}")
```
